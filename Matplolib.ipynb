```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0, 2, 10)
y = x**x
plt.figure(figsize=(12, 8))
plt.title('Fuigure 1')
plt.plot(x, y, c='red', lw=4, ls='--', label = 'quadratique')
plt.plot(x, x**3, label = 'cubique')
plt.xlabel('axe x')
plt.ylabel('axe y')
plt.legend()
plt.savefig('notre_figure.png')
plt.show() # Pour afficher le graphique
```

```python
plt.figure()
plt.subplot(2, 2, 1)
plt.plot(x, y)
plt.title('Graphique 1')
plt.subplot(2, 2, 2)
plt.plot(x, np.sin(x), label='sinus')
plt.plot(x, np.cos(x), label='cosinus')
plt.title('Graphique 2')
plt.legend()
```
